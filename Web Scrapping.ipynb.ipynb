```python
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By

from webdriver_manager.chrome import ChromeDriverManager
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
# driver.get("https://www.python.org/")
```

```python
info = driver.find_element(By.ID, "gsc_prf_i").text
Title = driver.find_element(By.ID, "gsc_a_t").text

print(info)
print('\n')
print(Title)
```

```python
show_more_button = driver.find_element(By.ID, "gsc_bpf_more")
show_more_button.click()
```

```python
title_tr = driver.find_elements(By.CLASS_NAME, "gsc_a_tr")

for tr in title_tr:
    print(tr.text)
    print('\n')
```

```python
ti = []

for t in title_tr:
    ti.append(t.text)
```

```python
import pandas as pd

df2 = pd.DataFrame(ti)
df2
```

```python
df2.to_csv("profile.csv", index = False)
```
